```python
import pandas as pd
import matplotlib.pyplot as plt

from pathlib import Path

filename = Path.cwd().parent / "data/Big Spender Portfolio Dec 2025.csv"
df = pd.read_csv(filename)

df['Symbol'] = df['Symbol'].str.split(':').str[-1]
df = df.drop(df.Symbol[df['Symbol'] == '$CASH'].index)

# ---- Chart: Frequency per Symbol ----
counts = df["Symbol"].value_counts()

plt.figure(figsize=(10, 5))
counts.plot(kind="bar")
plt.title("Number of Recommendations per Symbol")
plt.xlabel("Symbol")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
